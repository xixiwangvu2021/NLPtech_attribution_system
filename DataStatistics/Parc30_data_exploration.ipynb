```python
def remove_nested_attributions(value_list):
    return_value = ''
    for value in value_list:
        attr_elements_list = value.split('-')
        if len(attr_elements_list) > 1:
            if attr_elements_list[2] != 'NE':
                return_value = value.replace('-empty','').replace('-AT','').replace('-PD','').replace('-pd','')
    return return_value

df_train["col16"] = df_train["col11"]
df_train["col16"] = df_train.col16.str.split(' ')
df_train["col16"] = df_train['col16'].apply(remove_nested_attributions)
# df_train["col16"] = df_train.col16.str.join(' ')
df_train.head(20)
```

```python
# Function to plot most frequent terms.

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

def freq_words(words_df, terms=30):
    # selecting top n most frequent words
    d = words_df.nlargest(columns="counts", n=terms) 
    plt.figure(figsize=(20,5))
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=15)
    ax = sns.barplot(data=d, x="col7", y="counts")
    for item in ax.get_xticklabels():
        item.set_rotation(90)
    ax.set_xlabel('Word', fontsize=16)
    ax.set_ylabel('Count', fontsize=16)
    plt.show()
```
