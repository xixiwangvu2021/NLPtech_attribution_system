```python
# Function to plot most frequent terms.

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

def freq_words(words_df, terms=30):
    # selecting top n most frequent words
    d = words_df.nlargest(columns="counts", n=terms) 
    plt.figure(figsize=(20,5))
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=15)
    ax = sns.barplot(data=d, x="col7", y="counts")
    for item in ax.get_xticklabels():
        item.set_rotation(90)
    ax.set_xlabel('Word', fontsize=16)
    ax.set_ylabel('Count', fontsize=16)
    plt.show()
```

### Count of indirect quotations == without quotes
Mixed = two types of content label, within quotes and outside of the quotes

Select only content labels
Then group on content label numbers and join the tokens as one string.
And then apply a function to see if there are no quotes (indirect), and if there are quotes, if all tokens are within the quotes (direct), otherwise mixed.  

```python
# Function 'get_quote_type' unfortunately doesn't function 100% correct yet. 

def get_quote_type(text):
    """ This function figures out what quote type the CONTENT of a quote is. """
    quote_opening = "``"
    quote_closing = "''"
    text = text.strip()
    token_list = text.split(' ')
    
    quote_type = ''
    if quote_opening not in text and quote_closing not in text:
        quote_type = 'indirect'
    else:
        if text[0:2] == quote_opening and text[-2:] == quote_closing:
            quote_type = 'direct'

        quote_opening_count = 0
        quote_closing_count = 0
        
        # TODO, Instead of looping over the character, try looping over the token_list here.
        for index, character in enumerate(text[2:-1]):
            # If we find a quote closing marker and not directly a quote opening marker, then it's an mixed quote
            
            if text[index:index+2] == quote_opening:
                quote_opening_count += 1
            if text[index:index+2] == quote_closing:
                quote_closing_count += 1
                
            # quote_opening_count == quote_closing_count: to rule out cases like 'armed with `` assault weapons ''.' within a quote.
            if quote_opening_count == quote_closing_count and text[index:index+2] != quote_closing and text[index+3:index+5] == quote_opening:
                quote_type = 'mixed'
                break
    return quote_type                
```
